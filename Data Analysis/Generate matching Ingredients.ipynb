```python
title_no = list(dict(Counter(ting)).keys())
title_counter = list(dict(Counter(ting)).values())
title = citations.Title.values
```

```python
#3 variables
ftitle = []
stitle = []
sting = []
ting_comp = []

#Find duplicates in ting
for inx,ind in enumerate(title_counter):
    tileindx=ting.index(ting[inx],0)+1
    print(inx,ind)
    print(title_no[inx])
    if title_no[inx] not in ting_comp:
        ting_comp.append(title_no[inx])
    else:
        
        continue
    #print('tindx',tileindx)
    if ind == 1:
        continue
    for inn in range(ind-1):
        print('INN',inn)
        tileindx = ting.index(title_no[inx],tileindx)
        #print('TI',tileindx)
        ftitle.append(title[ting.index(title_no[inx],inx)])
        stitle.append(title[tileindx])
        sting.append(title_no[inx])
        tileindx+=1
```
